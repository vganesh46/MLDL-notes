###  Let’s start simple with a 1D function, y = f(x)

Say, you want to find x to minimise the function. What do you do? 

Solve for its derivate function f’(x) = 0. This gives you some values of x. Evaluate y at these points (these are called critical points) and see where y attains the minimum, right? (But, remember f’(x) = 0 applies for those points where y attains the maximum too)

Another approach to minimise the function is to use an iterative algorithm and avoid solving the equation f’(x) = 0 . Choose an arbitrary point x and the derivative function f’(x) at x gives you the rate at which y changes w.r.t x. So, you go along its negative direction for the new x, keep calculating f’(x) and stop at a point where f’(x) ~ 0 to find the minimum. This approach is called Gradient Descent.

                                                x’ = x - E*(f’(x))                
where E can be considered as the size of step by which we move x to the new point.

### Let's add some complexity with a multidimensional function w = f(x)
i.e., x is a point in multidimensional space say (x,y,z).

The above equation modifies to:

                                            x’ = x - E*(grad(f(x))

Where grad(f(x)) denotes the gradient of function f. Gradient of function  is basically a vector containing all the partial derivates.  But, why is negative direction of gradient the preferred direction to minimise f?

To minimise f, we find the direction in which f decreases the fastest and  this is in the negative direction of Gradient. This is not totally intuitive. Let me explain this the following way.

Each component of gradient i.e., the partial derivate tells us how fast the function is changing w.r.t. to standard directions (x,y,z) . How fast the function changes in an arbitrary direction is the dot product(grad(f(x).v) where v is a unit vector in that direction. And, so the function increases the fastest in the direction of gradient as that maximises the dot product. In other words, you go the opposite direction to find the minimum of function. 

Here too, you can take the iterative gradient descent approach or solve the equation grad(f(x)) = 0 to arrive a x that minimises the function. The thing to keep in mind is we will not always end up with global minimum. Depending on the startint arbitary point and the step size we use to move, we can end up in a different local minimum.
