While ROC curve represents the relation between Sensitivity (Recall) and Specificity at varying classification thresholds of a classifier, Precision-Recall curve represents the relation between Precision and Recall.So, in a setting, if we are more concerned about the precision, we would expect the Precision-Recall (PR) curve to help us better in choosing the better classifier.

If you recall, Precision = P (y = 1 | yhat = 1) and hence will vary in different populations with different P (y = 1). However, PR-curve is very useful in cases if we care about one population with known background probability and where positive class is more interesting than the negative class (this is unlike ROC curve which deals with negative class).

### ROC curve vs PR Curve

Compare the two curves for two classifier algorithms below. If you see, both have a very similar ROC curve, but algorithm 2 performs better in PR curve. This can happen when there are large number of negative examples compared with positive examples in the dataset.

Consider the example of a fraud detection problem (large negative examples). To compare any two classifiers performance, the difference between a good vs bad classifier in this scenario is more apparent when we use a PR curve.

    Algorithm 1: 90 actual fraud out of 100 identified fraud
    Algorithm 2: 90 actual fraud out of 1000 identified fraud.

    In the ROC curve context, the difference (FPR) between the two algorithms is insignificant.

    Algorithm 1: TPR=90/100=0.9, FPR= 10/1,999,900=0.00000500025
    Algorithm 2: TPR=90/100=0.9, FPR=910/1,999,900=0.00045502275

    The FPR difference is 0.0004500225

    In the PR Curve context, the difference (precision) is very significant.

    Algorithm 1: precision=0.9, recall=0.9
    Algorithm 2: Precision=90/1000=0.09, recall= 0.9
    Precision difference= 0.81
